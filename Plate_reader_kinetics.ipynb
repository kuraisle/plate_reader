```python
#@title Press play on this cell to start
import numpy as np
import pandas as pd
from google.colab import files
import io
import ipywidgets as widgets
from IPython.display import display
```

```python
#@title File upload
#@markdown Upload a csv of your data here. Please remove the descriptive header and keep only the rows below the "Well, Content.." column titles.
uploaded = files.upload()

file_csv = pd.read_csv(io.BytesIO(list(uploaded.values())[0]), header = [0,1], index_col = 0)
file_csv = file_csv.drop('Content', axis = 1)
times = [x[1] for x in file_csv.columns]
measures = [x[0] for x in file_csv.columns]
wells = file_csv.index
well_rows = [x[0] for x in wells]
well_cols = [int(x[1:]) for x in wells]
```

```python
#@title Select conditions
#@markdown Choose what conditions vary across your wells

#@markdown Press "Run Interact" to update the list of conditions
conditions = []
for i in range(1,11):
  conditions.append({
      'active': False,
      'description':'',
      'discrete':'Discrete'
  })

def edit_condition(condition_num, active, description, discrete):
  conditions[condition_num-1] = {'active': active, 'description':description, 'discrete':discrete}
  for i in range(0,10):
    a, des, dis = conditions[i].values()
    if a:
      print(f"Condition {i+1}:")
      print(f"{des} ({dis})")

w = widgets.interact_manual(edit_condition, condition_num = list(range(1,11)), active = False, description = '', discrete = ['Discrete', 'Continuous'])
```

```python
#@title Set your condition parameters
#@markdown Press play. If your conditions are discrete, you need to set some options. If they're continuous, you need to set units.

#@markdown To set multiple options, enter a list separated by a comma and a space
condition_pars = dict()

for condition in conditions:
  if condition['active']:
    if condition['discrete'] == 'Discrete':
      condition_pars[condition['description']] = {'discrete': True, 'val': []}
    else:
      condition_pars[condition['description']] = {'discrete': False, 'val': ''}

cond_par_controls = dict()

def par_updater(parameter):
  if condition_pars[parameter]['discrete']:
    def fun(options):
      condition_pars[parameter]['val'] = options.split(', ')
  else:
    def fun(unit):
      condition_pars[parameter]['val'] = unit
  return fun

for par in condition_pars.keys():
  if condition_pars[par]['discrete']:
    cond_par_controls[par] = widgets.interactive(par_updater(par), options = '')
  else:
    cond_par_controls[par] = widgets.interactive(par_updater(par), unit = '')


for control_name, control in cond_par_controls.items():
  print(control_name)
  display(control)
```

```python
#@title Set the options for your wells

well_pars = dict()

for well in wells:
  well_pars[well] = dict()
  for condition_name, condition in condition_pars.items():
    if condition['discrete']:
      well_pars[well][condition_name] = condition['val'][0]
    else:
      well_pars[well][condition_name] = 0

#def well_pars_updater(well):
#  for par in condition_pars:
    

well_pars
```
